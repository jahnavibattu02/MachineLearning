### Import Libraries and Loading the data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Get the data
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
#The data set is presented in dictionary form:
data.keys()
```

```python
print(data['DESCR'])
```

```python
data['feature_names']
```

```python
#Create a DataFrame named as dataset
dataset = pd.DataFrame(data['data'], columns=data['feature_names'])
dataset.head()
```

```python
dataset.info()
```

```python
#Having look at Target column
data['target']
```

```python
df_target = pd.DataFrame(data['target'], columns=['Cancer'])
```

```python
dataset.head()
```

### Splitting the data into Train and Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataset, np.ravel(df_target), test_size=0.20, random_state=101)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Generalize the Model

```python
from sklearn.svm import SVC
```

```python
model = SVC() #Initialize the model
```

```python
model.fit(X_train, y_train) # Train the model with train data
```

### Predictions and Evaluations.

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(accuracy_score(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```

### Hyperparameter Tuning Using GridSearch

```python
param_grid = {'C':[0.1, 1, 10, 100, 1000],
              'gamma':[1, 0.1, 0.01, 0.001, 0.0001],
              'kernel':['rbf','sigmoid','linear']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid, refit=True, verbose=3)
```

```python
grid.fit(X_train, y_train) #Fitting the model
```

```python
grid.best_params_ #Return the best param values
```

```python
grid.best_estimator_ # Returns best estimator.
```

### Model Evaluation After Tuning the parameters

```python
grid_predictions = grid.predict(X_test)
```

```python
print(accuracy_score(y_test, grid_predictions))
```

```python
print(confusion_matrix(y_test, grid_predictions))
```

```python
print(classification_report(y_test, grid_predictions))
```
